```python
# removes punctuation
# takes a string only tho 
def remove_punctuation(text):
    #pat = re.compile(ur"[\p{P}\p{S}]+")
    pat = re.compile(ur"[\p{P}\p{S}](?<!-)+")
    return pat.sub("", text)
```
